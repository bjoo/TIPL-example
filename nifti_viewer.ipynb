# NIFTI File Viewer
## Load a NIFTI file

```c++
#include "TIPL/tipl.hpp" // git clone http://github.com/frankyeh/TIPL/
using namespace tipl;
image<3> I;
I.load_from_file<io::nifti>("data/mni_icbm152_t1.nii");
```

### Create a slider bar

```c++
#include "xwidgets/xslider.hpp"
#include "xcpp/xdisplay.hpp"
// now create a slider bar for slice pos
auto slider = xw::slider<int>::initialize().min(0).max(I.depth()-1).value(I.depth()/2).finalize();
xeus::xguid display_id;
XOBSERVE(slider, value,[&](const auto& s) {
    xcpp::display(I.slice_at(s.value),display_id,true);
});
```

### Visualize results

```c++
display_id = xeus::new_xguid();
xcpp::display(I,display_id,false);
slider.display();
```
