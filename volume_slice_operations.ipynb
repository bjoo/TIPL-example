# Volume and Slice operations

This example shows how to use volume operators

```c++
#include "TIPL/tipl.hpp" // git clone http://github.com/frankyeh/TIPL/
using namespace tipl;
```

```c++
image<3> I;
I.load_from_file<tipl::io::nifti>("data/mni_icbm152_t1.nii");
I
```

## Volume Operations


```c++
swap_xy(I)
```

```c++
flip_x(I)
```

```c++
crop(I,v(30,30,0),v(80,80,80))
```

## Slice Operation

```c++
I.load_from_file<tipl::io::nifti>("data/mni_icbm152_t1.nii");
I.slice_at(10)
```

```c++
image<float,2> slice;
volume2slice(I,slice,0,100); // 0: saggital slice at 100
slice
```

```c++
volume2slice(I,slice,1,100); // 0: coronal slice at 100
slice
```
